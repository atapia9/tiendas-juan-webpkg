## 6) Marco de decisión (scorecard)
Se pondera cada tienda con 3 factores:

- **Facturación** (más es mejor) — peso 0.5
- **Calificación** (más es mejor) — peso 0.3
- **Costo de envío** (menos es mejor) — peso 0.2

Normalizamos cada métrica (0–1) y calculamos un puntaje compuesto. La **candidata a vender** será la de **menor puntaje**.

```python
# Normalización [0,1]
def minmax(s, invert=False):
    s = s.astype(float)
    m, M = s.min(), s.max()
    if M == m:
        return pd.Series(np.ones(len(s))*0.5, index=s.index)
    z = (s - m) / (M - m)
    return (1 - z) if invert else z

sc = facturacion.merge(prom_eval, on="tienda").merge(envio_prom, on="tienda")
sc["fact_n"] = minmax(sc["Facturación Total"])
sc["calif_n"] = minmax(sc["Calificación Promedio"])
sc["envio_n"] = minmax(sc["Costo Promedio de Envío"], invert=True)

w_fact, w_calif, w_envio = 0.5, 0.3, 0.2
sc["score"] = w_fact*sc["fact_n"] + w_calif*sc["calif_n"] + w_envio*sc["envio_n"]
sc_sorted = sc.sort_values("score", ascending=False)  # mayor score = mejor tienda
sc_sorted[["tienda","Facturación Total","Calificación Promedio","Costo Promedio de Envío","fact_n","calif_n","envio_n","score"]]
```

# ETL mínimo: datos crudos → `df`
Lee todos los CSV/XLSX de `../data/raw/`, normaliza columnas, tipifica numéricos, une y deduplica.


```python

# %%
import os, glob, pandas as pd, numpy as np

RAW_DIR = os.path.normpath(os.path.join("..", "data", "raw"))
INTERMEDIATE_DIR = os.path.normpath(os.path.join("..", "data"))
os.makedirs(RAW_DIR, exist_ok=True)

print("📂 Carpeta de entrada:", RAW_DIR)

RENAME_RULES = {
    r"tienda|sucursal|store|shop": "tienda",
    r"categor[ií]a|category": "categoria",
    r"producto|product|item": "producto",
    r"facturaci[oó]n|revenue|ventas|importe|monto|total": "facturacion",
    r"calificaci[oó]n|evaluaci[oó]n|rating|score": "evaluacion",
    r"(costo.*env[ií]o)|(shipping.*cost)|(env[ií]o)": "costo_envio",
    r"cantidad|unidades|qty|quantity|ventas_n|n_ventas": "cantidad",
    r"^lat(itud)?$|latitude": "lat",
    r"^lon(gitud)?$|long|longitude": "lon"
}
def normalize_columns(cols):
    out = []
    import re
    for c in cols:
        lc = c.strip().lower().replace(" ", "_")
        newc = lc
        for pat, target in RENAME_RULES.items():
            if re.search(pat, lc, flags=re.IGNORECASE):
                newc = target
                break
        out.append(newc)
    return out
def coerce_numeric(s):
    if s.dtype.kind in "biufc":
        return s
    return (s.astype(str)
              .str.replace(r"[^\d,\.\-]", "", regex=True)
              .str.replace(",", ".", regex=False)
              .replace({"": np.nan, ".": np.nan})
              .astype(float))

files = glob.glob(os.path.join(RAW_DIR, "**", "*.csv"), recursive=True) + \
        glob.glob(os.path.join(RAW_DIR, "**", "*.xlsx"), recursive=True) + \
        glob.glob(os.path.join(RAW_DIR, "**", "*.xls"), recursive=True)

if not files:
    print("⚠️ No se encontraron archivos en", RAW_DIR, " — coloca aquí tus CSV/XLSX y vuelve a ejecutar.")
else:
    print(f"Se encontraron {len(files)} archivos.")
frames = []
for path in files:
    try:
        if path.lower().endswith(".csv"):
            df_i = pd.read_csv(path, encoding="utf-8", engine="python")
        else:
            df_i = pd.read_excel(path, engine="openpyxl")
        df_i.columns = normalize_columns(df_i.columns)
        for cand in ["facturacion", "evaluacion", "costo_envio", "cantidad"]:
            if cand in df_i.columns:
                df_i[cand] = coerce_numeric(df_i[cand])
        frames.append(df_i)
        print("✔️ Cargado:", os.path.basename(path), df_i.shape)
    except Exception as e:
        print("❌ Error con", path, "→", e)
if frames:
    df = pd.concat(frames, ignore_index=True, sort=False).drop_duplicates()
    inter_csv = os.path.join(INTERMEDIATE_DIR, "df_intermedio.csv")
    df.to_csv(inter_csv, index=False, encoding="utf-8")
    print("🗂️ Intermedio guardado en:", inter_csv)
    display(df.head(10))
else:
    print("⚠️ No se creó el DataFrame `df` porque no hubo archivos válidos.")
```

```python

# %% [markdown]
# # Exportador de métricas para el sitio (GitHub Pages)

# %%
import json, os, pandas as pd
COLUMN_SYNONYMS = {
    "store": ["tienda", "store", "sucursal", "shop"],
    "category": ["categoria", "categoría", "category"],
    "product": ["producto", "product", "item"],
    "revenue": ["facturacion", "facturación", "revenue", "ventas_mxn", "monto", "total", "importe"],
    "rating": ["evaluacion", "evaluación", "rating", "calificacion", "calificación", "score"],
    "ship_cost": ["costo_envio", "costo_de_envio", "shipping_cost", "envio", "costo_logistico", "costo_logístico"],
    "qty": ["cantidad", "unidades", "qty", "quantity", "ventas", "n_ventas"]
}
def find_col(df, keys):
    cols = {c.lower(): c for c in df.columns}
    for k in keys:
        k = k.lower()
        if k in cols: 
            return cols[k]
    for k in keys:
        for lc, orig in cols.items():
            if k in lc:
                return orig
    return None
def resolve_columns(df):
    mapping = {}
    for std, candidates in COLUMN_SYNONYMS.items():
        col = find_col(df, candidates)
        mapping[std] = col
    missing = [k for k, v in mapping.items() if v is None and k in ("store","revenue","rating","ship_cost","qty","category","product")]
    if len(missing) > 0:
        print("⚠️ Columnas no detectadas automáticamente:", missing)
    return mapping
try:
    df  # noqa: F821
except NameError:
    raise RuntimeError("No se encontró el DataFrame 'df'. Ejecuta el ETL primero.")
mapping = resolve_columns(df)
S = mapping["store"]; C = mapping["category"]; P = mapping["product"]
R = mapping["revenue"]; Q = mapping["qty"]; RT = mapping["rating"]; SC = mapping["ship_cost"]
d = df.copy()
stores = sorted(d[S].dropna().unique().tolist()) if S else []
fact_by_store = d.groupby(S)[R].sum().reindex(stores).round(2).tolist() if (S and R) else [None]*len(stores)
eval_by_store = d.groupby(S)[RT].mean().reindex(stores).round(2).tolist() if (S and RT) else [None]*len(stores)
ship_by_store = d.groupby(S)[SC].mean().reindex(stores).round(2).tolist() if (S and SC) else [None]*len(stores)
categorias_top = {}
if S and C:
    if Q and Q in d.columns:
        tmp = d.groupby([S, C])[Q].sum().reset_index()
        qcol = Q
    else:
        tmp = d.groupby([S, C]).size().reset_index(name="__cnt__")
        qcol = "__cnt__"
    for st in stores:
        sub = tmp[tmp[S]==st].sort_values(qcol, ascending=False).head(5)
        categorias_top[st] = {row[C]: int(row[qcol]) for _, row in sub.iterrows()}
productos_extremos = {}
if S and P:
    if Q and Q in d.columns:
        tmp = d.groupby([S, P])[Q].sum().reset_index()
        qcol = Q
    else:
        tmp = d.groupby([S, P]).size().reset_index(name="__cnt__")
        qcol = "__cnt__"
    for st in stores:
        sub = tmp[tmp[S]==st].sort_values(qcol, ascending=False)
        top = sub.head(3)[P].tolist()
        bottom = sub.tail(2)[P].tolist() if len(sub)>=2 else sub.tail(1)[P].tolist()
        productos_extremos[st] = {"mas_vendidos": top, "menos_vendidos": bottom}
metrics = {
    "updated_at": pd.Timestamp.utcnow().isoformat(),
    "stores": stores,
    "facturacion_total": fact_by_store,
    "promedio_evaluacion": eval_by_store,
    "costo_envio_promedio": ship_by_store,
    "categorias_top": categorias_top,
    "productos_extremos": productos_extremos
}
out_path = os.path.normpath(os.path.join("..", "data", "metrics.json"))
os.makedirs(os.path.dirname(out_path), exist_ok=True)
with open(out_path, "w", encoding="utf-8") as f:
    json.dump(metrics, f, ensure_ascii=False, indent=2)
print("✅ Métricas exportadas a:", out_path)
pd.DataFrame(metrics).to_csv(os.path.join(os.path.dirname(out_path), "metrics_flat.csv"), index=False, encoding="utf-8")
```

# Análisis geográfico (opcional)
Genera dispersión, heatmap (hexbin) y un **mapa Folium** con *MarkerCluster* y *HeatMap* a partir de columnas `lat`/`lon` (o sinónimos).


```python

# %%
import os, math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

LAT_CANDIDATES = ["lat", "latitude", "latitud"]
LON_CANDIDATES = ["lon", "long", "longitude", "longitud"]
STORE_CANDS   = ["tienda", "store", "sucursal", "shop"]
REV_CANDS     = ["facturacion", "facturación", "revenue", "ventas_mxn", "monto", "total", "importe"]
RATE_CANDS    = ["evaluacion", "evaluación", "rating", "calificacion", "calificación", "score"]

def find_first(cols, cands):
    lower = {c.lower(): c for c in cols}
    for k in cands:
        if k in lower: 
            return lower[k]
    for k in cands:
        for lc, orig in lower.items():
            if k in lc:
                return orig
    return None

assert 'df' in globals(), "Se requiere el DataFrame `df` (ejecuta el ETL)."

lat_col = find_first(df.columns, LAT_CANDIDATES)
lon_col = find_first(df.columns, LON_CANDIDATES)
store_col = find_first(df.columns, STORE_CANDS)
rev_col = find_first(df.columns, REV_CANDS)
rate_col = find_first(df.columns, RATE_CANDS)

if not lat_col or not lon_col:
    raise RuntimeError("No se detectaron columnas de latitud/longitud. Ajusta nombres o agrega columnas `lat` y `lon`.")

geo = df[[c for c in [store_col, lat_col, lon_col, rev_col, rate_col] if c in df.columns]].copy()
geo = geo.dropna(subset=[lat_col, lon_col])
geo[lat_col] = pd.to_numeric(geo[lat_col], errors="coerce")
geo[lon_col] = pd.to_numeric(geo[lon_col], errors="coerce")
geo = geo.dropna(subset=[lat_col, lon_col])

assets_dir = os.path.normpath(os.path.join("..", "assets"))
os.makedirs(assets_dir, exist_ok=True)

# Dispersión por tienda
plt.figure(figsize=(7,6))
if store_col and store_col in geo.columns:
    groups = geo.groupby(store_col)
    for name, g in groups:
        plt.scatter(g[lon_col], g[lat_col], s=14, alpha=0.6, label=str(name))
    plt.legend(title="Tienda", fontsize=8)
else:
    plt.scatter(geo[lon_col], geo[lat_col], s=14, alpha=0.6)
plt.title("Dispersión geográfica de transacciones")
plt.xlabel("Longitud"); plt.ylabel("Latitud")
scatter_path = os.path.join(assets_dir, "geo_scatter.png")
plt.tight_layout(); plt.savefig(scatter_path, dpi=140); plt.close()
print("📈 Dispersión guardada:", scatter_path)

# Heatmap hexbin
plt.figure(figsize=(7,6))
hb = plt.hexbin(geo[lon_col], geo[lat_col], gridsize=40, cmap="viridis", mincnt=1)
cb = plt.colorbar(hb); cb.set_label("Densidad")
plt.title("Mapa de calor (hexbin) de densidad de ventas")
plt.xlabel("Longitud"); plt.ylabel("Latitud")
hex_path = os.path.join(assets_dir, "geo_heatmap_hexbin.png")
plt.tight_layout(); plt.savefig(hex_path, dpi=140); plt.close()
print("🔥 Heatmap hexbin guardado:", hex_path)

# Folium interactivo
try:
    import folium
    from folium.plugins import MarkerCluster, HeatMap
    center_lat = float(geo[lat_col].median())
    center_lon = float(geo[lon_col].median())
    m = folium.Map(location=[center_lat, center_lon], zoom_start=6, tiles="CartoDB positron")
    mc = MarkerCluster()
    if store_col and store_col in geo.columns:
        for _, r in geo.iterrows():
            popup = []
            if store_col: popup.append(f"<b>Tienda:</b> {r.get(store_col, '')}")
            if rev_col and rev_col in geo.columns and not pd.isna(r.get(rev_col)):
                try: popup.append(f"<b>Ingreso:</b> {float(r.get(rev_col)):.2f}")
                except: popup.append(f"<b>Ingreso:</b> {r.get(rev_col)}")
            if rate_col and rate_col in geo.columns and not pd.isna(r.get(rate_col)):
                try: popup.append(f"<b>Evaluación:</b> {float(r.get(rate_col)):.2f}")
                except: popup.append(f"<b>Evaluación:</b> {r.get(rate_col)}")
            folium.Marker([r[lat_col], r[lon_col]],
                          popup=folium.Popup("<br/>".join(popup) if popup else "Transacción", max_width=300),
                          icon=folium.Icon(color="blue", icon="shopping-cart", prefix="fa")).add_to(mc)
    else:
        for _, r in geo.iterrows():
            folium.Marker([r[lat_col], r[lon_col]]).add_to(mc)
    mc.add_to(m)
    if rev_col and rev_col in geo.columns:
        heat_data = geo[[lat_col, lon_col, rev_col]].dropna().values.tolist()
        vals = geo[rev_col].dropna().astype(float).values if len(geo[rev_col].dropna())>0 else []
        if len(vals) > 0 and np.nanmax(vals) > 0:
            maxv = np.nanpercentile(vals, 95)
            if maxv and maxv > 0:
                heat_data = [[lt, ln, min(float(rv)/maxv, 1.0)] for lt, ln, rv in heat_data]
        HeatMap(heat_data, radius=16, blur=22, min_opacity=0.3).add_to(m)
    else:
        HeatMap(geo[[lat_col, lon_col]].dropna().values.tolist(), radius=16, blur=22, min_opacity=0.3).add_to(m)
    folium_path = os.path.join(assets_dir, "ventas_mapa.html")
    m.save(folium_path)
    print("🗺️ Mapa Folium guardado:", folium_path)
except Exception as e:
    print("⚠️ Folium no disponible o error al generar mapa:", e)

# Agregados espaciales por celdas (binning simple)
geo["_lat_bin"] = geo[lat_col].round(1)
geo["_lon_bin"] = geo[lon_col].round(1)
agg = geo.groupby(["_lat_bin","_lon_bin"]).size().reset_index(name="n")
if rev_col and rev_col in geo.columns:
    agg["revenue_sum"] = geo.groupby(["_lat_bin","_lon_bin"])[rev_col].sum().values
if rate_col and rate_col in geo.columns:
    agg["rating_mean"] = geo.groupby(["_lat_bin","_lon_bin"])[rate_col].mean().values
agg = agg.rename(columns={"_lat_bin":"lat_bin","_lon_bin":"lon_bin"})
agg_path = os.path.join(assets_dir, "geo_bins.csv")
agg.to_csv(agg_path, index=False, encoding="utf-8")
print("📊 Agregados espaciales guardados:", agg_path)
```
